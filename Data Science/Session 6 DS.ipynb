```python
label = ["A" , "B" , "C" , "D"]
y     = [25  , 25  , 15  , 35 ]
plt.pie(y , labels=label , autopct = '%2.f%%' ,colors = ["red" , "green" , "blue" , "black"])
plt.show()
```

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct

plt.pie(y, labels=label, autopct=make_autopct(y))
plt.show()
```
