```python
Q1 = df["price"].quantile(.25)
print("Q1",Q1)
Q3 = df["price"].quantile(.75)
print("Q3",Q3)
IQR = Q3 - Q1
print("IQR", IQR)
```

```python
lower_value = Q1 -  (1.5 * IQR)
print("lower_value",lower_value)
upper_value = Q3 +  (1.5 * IQR)
print("upper_value", upper_value)
```

```python
df[["price"]].describe()
```

```python
df_without_out = df[(df["price"] >= lower_value) & (df["price"] <= upper_value)]   # & and ,  | or 
```

```python
df_without_out["price"].max()
```

```python
df[(df["price"] < lower_value) | (df["price"] > upper_value)]["price"].max()
```
